```python
# Import modules
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
train = pd.read_csv('../data/train.csv', index_col=0)
train_clean = pd.read_csv('../data/train_clean.csv', index_col=0)
test = pd.read_csv('../data/test.csv', index_col=0)
```

```python
#Create building age and remodel age columns
building_age = 2019-train_clean[['YearBuilt']]
re_age = 2019-train_clean[['YearRemodAdd']]
building_age[np.isnan(building_age)] = 0
re_age[np.isnan(re_age)] = 0
train_clean = train.assign(building_age=building_age, re_age=re_age)
```

```python
#normalize
nreage = (train_clean.re_age - train_clean.re_age.mean())/train_clean.re_age.std()
nbage = (train_clean.building_age - train_clean.building_age.mean())/train_clean.building_age.std()
ngla = (train_clean.GrLivArea - train_clean.GrLivArea.mean())/train_clean.GrLivArea.std()
noq = (train_clean.OverallQual - train_clean.OverallQual.mean())/train_clean.OverallQual.std()
nla = (train_clean.LotArea - train_clean.LotArea.mean())/train_clean.LotArea.std()
nor_sale = (train_clean[['SalePrice']] - train_clean[['SalePrice']].mean())/train_clean[['SalePrice']].std()
```

```python
#add normalize feature to dataframe
train_clean = train.assign(nreage=nreage, nbage=nbage, ngla=ngla, noq=noq, nla=nla, nor_sale=nor_sale)
```

```python
#normalize re_age
nlm = LinearRegression()
nlm.fit(nor_sale, nreage)
print('intercept %.2f' %(nlm.intercept_))
print('slope %.6f' %(nlm.coef_))
print('R-Squared %.2f' %(nlm.score(nor_sale,nreage)))
print('RSS %.2f' %(np.sum((nreage - nlm.predict(nor_sale))**2)))
train_clean.plot.scatter(x='nreage', y='nor_sale')
```

```python
#normalize building_age
nlm1 = LinearRegression()
nlm1.fit(nor_sale, nbage)
print('intercept %.2f' %(nlm1.intercept_))
print('slope %.6f' %(nlm1.coef_))
print('R-Squared %.2f' %(nlm1.score(nor_sale,nbage)))
print('RSS %.2f' %(np.sum((nbage - nlm1.predict(nor_sale))**2)))
train_clean.plot.scatter(x='nbage', y='nor_sale')
```

```python
#normalize GrLivArea
nlm2 = LinearRegression()
nlm2.fit(nor_sale, ngla)
print('intercept %.2f' %(nlm2.intercept_))
print('slope %.6f' %(nlm2.coef_))
print('R-Squared %.2f' %(nlm2.score(nor_sale,ngla)))
print('RSS %.2f' %(np.sum((ngla - nlm2.predict(nor_sale))**2)))
train_clean.plot.scatter(x='ngla', y='nor_sale')
```

```python
#normalize OverallQual
nlm3 = LinearRegression()
nlm3.fit(nor_sale, noq)
print('intercept %.2f' %(nlm3.intercept_))
print('slope %.5f' %(nlm3.coef_))
print('R-Squared %.2f' %(nlm3.score(nor_sale,noq)))
print('RSS %.2f' %(np.sum((noq - nlm3.predict(nor_sale))**2)))
train_clean.plot.scatter(x='noq', y='nor_sale')
```

```python
#normalize LotArea
nlm4 = LinearRegression()
nlm4.fit(nor_sale, nla)
print('intercept %.2f' %(nlm4.intercept_))
print('slope %.6f' %(nlm4.coef_))
print('R-Squared %.2f' %(nlm4.score(nor_sale,nla)))
print('RSS %.2f' %(np.sum((nla - nlm4.predict(nor_sale))**2)))
train_clean.plot.scatter(x='nla', y='nor_sale')
```
